### 1. When do the first and last flights leave each day?

```python

import numpy as np
flights.groupby(['day']).agg({'dep_time': [np.min,np.max]})
```

```python
def first_last(df):
    return df.iloc[[0, -1]]
grouped_by_month_day = flights.groupby(['month', 'day']).apply(first_last)

grouped_by_month_day
```

### 2. When there is a missing value in dep_time then this is an indication of canceled flight. Find the number of cancelled flights for each (carrier, origin) combination.



```python
df=flights[flights.dep_time.isnull()][['carrier', 'origin']]
len(df)
```

### 3. Which carrier has the worst dep_delays?

```python
#flights.groupby(['carrier', 'origin']).agg({'dep_time': 'count'})

old=flights.groupby(['carrier']).agg({'dep_delay': 'max'})
old[old['dep_delay']==old['dep_delay'].max()]
```

### 4. Which plane (tailnum) has the worst on-time record?

```python
#Which plane (tailnum) has the worst on-time record?
new = flights.groupby(['tailnum']).agg({'arr_delay': 'max'})
new[new['arr_delay']==new['arr_delay'].max()]
```

### 5. For each plane, count the number of flights before the first delay of greater than 1 hour.

```python
count=0
for name, group in flights.groupby(['day', 'tailnum']):
     for index, row in group.iterrows():
                 if row['dep_delay']  < 60:
                     count +=1
                 else:
                     print('index, day, count',(row['tailnum'],row['day'], count))
                     count = 0
                     break
```
