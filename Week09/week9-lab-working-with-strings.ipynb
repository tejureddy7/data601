### 1.Find all the answers with 'jupyter' mentioned in it.

```python
pattern = re.compile(r'Jupyter')
df['mentioned_jupyter'] = [True if pattern.search(str(s))!=None  else False for s in df['Q12_Part_4_TEXT']]
```

```python
 df[df['mentioned_jupyter'] == True]['Q12_Part_4_TEXT']
```

### 2.Find all the participant who are Python users.

```python
pattern = re.compile(r'python|Python')
df['mentioned_jupyter'] = [True if pattern.search(str(s))!=None  else False for s in df['Q12_Part_4_TEXT']]
```

```python
 df[df['mentioned_jupyter'] == True]['Q12_Part_4_TEXT']
```

### 3.Find all R users.

```python
pattern = re.compile(' Rstudio|Rstudio|^R, | R,| R', re.IGNORECASE)
#pattern.search(r'^r$|rstudio, re.I) r.*
df['mentioned_jupyter'] = [True if pattern.search(str(s))!=None  else False for s in df['Q12_Part_4_TEXT']]
```

```python
 df[df['mentioned_jupyter'] == True]['Q12_Part_4_TEXT']
```
