````python
import random 
import matplotlib.pyplot as plt
import statistics

trial_count = 500

def fair_die_simulation(trials):
    """
    returns an array of randomly numbers between 1 and 6. 
    """
    options = [1,2,3,4,5,6]
    simulations = []
    count=0
    while count < trials:
        simulations.append(random.choice(options))
        count +=1   
    return simulations
    
fair_die_simulation(500)
````

````python
import numpy 

def biased_die_simulations(trials, weights=[0.25, 0.15, 0.15, 0.15, 0.15, 0.15]):
    """
    returns an array of randomly numbers between 1 and 6 with the probabilty of assosiated weight. 
    for instance: for weights: [0.25, 0.15, 0.15, 0.15, 0.15, 0.15]
    1 has 0.25 chance to be selected, 2 has 0.15 chance and so on...
    """
    count =0
    simulations =[]
    while count < 500:
        p =weights
        simulations.append(numpy.random.choice(numpy.arange(1, 7), p=weights))
        count +=1
    return simulations
   
    
biased_die_simulations(500)
````

````python
#calculating mean for fair die 
def fair_mean_list():
    fair_mean_list = []
    for i in range(500):
        fair_mean_list.append(statistics.mean(fair_die_simulation(500)))
    return fair_mean_list

fair_mean_list()
                         
````

````python
#calculating mean for biased die
def biased_mean_lists():
    biased_mean_list = []
    for i in range(500):
        biased_mean_list.append(sum(biased_die_simulations(500))/len(biased_die_simulations(500)))
    return biased_mean_list

biased_mean_lists()
````

````python
#biased to store biased simulation
biased= biased_die_simulations(trial_count)
#fair to store fair die simulation
fair = fair_die_simulation(trial_count)

#biased and unbiased mean list
biased_mean_list = biased_mean_lists()
unbiased_mean_list = fair_mean_list()
````
